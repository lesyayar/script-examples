### Задание 2. Использование функций

Напишите функцию, которая **рекурсивно** вычисляет сумму элементов одномерного случайно заданного массива numpy. Продемонстрируйте работу функции на примерах.

```python
from random import randint
import numpy as np


def sum_elements_one_arr(n,sum_el = 0):
    if len(n) == 0:
        return 0
    else:
        sum_el += n[0] + sum_elements_one_arr(n[1:])
        return sum_el


N = int(input("Введите число элементов одномерного массива: "))
massiv = [randint(-10,10) for i in range(N)]
print(f"Получили случайно заданный массив: {massiv}")
sum_res = sum_elements_one_arr(massiv)
print(f"Сумма элементов = {sum_res}")
```

Первый вариант - max-heap, второй - min-heap. Так как двоичная куча - это законченное дерево, ее легко представить в виде массива. Если родительский узел хранится в индексе i: левый дочерний элемент вычисляется как 2i+1, а правый - как 2i+2(при условии, что индексирование начинается с 0).

### Задание 4. Использование алгоритмов сортировки

Реализовать функцию сортировки по алгоритму из Задания 3 с возможностью сортирования как по убыванию, так и по возрастанию. Необходимо продемонстрировать работу функции сортировки на различных примерах. Убедиться, что всё работает правильно, сравнив свои результаты с результатами стандартной функции sorted.

```python
#Пример работы алгоритма сортировки по возрастанию
import random

def HeapSort(lst, order): #основа
    for i in range(len(lst), -1, -1):
        heapify(lst, i, len(lst), order)
    for i in range(len(lst) - 1, 0, -1):
        lst[i], lst[0] = lst[0], lst[i]
        heapify(lst, 0, i, order)
    return lst


def heapify(lst, index, size, order): #просеивание
    maxi = index
    left = (2 * index) + 1 #левая дочерняя ветвь
    right = (2 * index) + 2 #правая дочерняя ветвь
    if order == "0":
        if left < size and lst[left] > lst[maxi]:
            maxi = left
        if right < size and lst[right] > lst[maxi]:
            maxi = right
        if maxi != index:
            lst[index], lst[maxi] = lst[maxi], lst[index]
            return heapify(lst, maxi, size, order)
    elif order == "1":
        if left < size and lst[left] < lst[maxi]:
            maxi = left
        if right < size and lst[right] < lst[maxi]:
            maxi = right
        if maxi != index:
            lst[index], lst[maxi] = lst[maxi], lst[index]
            return heapify(lst, maxi, size, order)


l = list(range(-10,10))
a = random.choices(l, k = 10)
odr = input("Выберите тип сортировки: по возрастанию (0) или по убыванию (1): ")
print(*a)
a_sort = HeapSort(a, odr)

if odr == "0":
    a_sort_python = sorted(a, reverse = False) 
elif odr == "1":
    a_sort_python = sorted(a, reverse = True) 
    
print(*a_sort)    
print(a_sort == a_sort_python)
```

### Задание 5. Поиск элементов
На вход программе в первой строке подаётся слово из M букв, а во второй строке - текст из N символов: букв и пробелов, N>M. 

Напишите функцию, которая определит позиции (в 0-индексации) всех вхождений заданного слова в текст. Для решения этой задачи необходимо построить хеш-таблицу (словарь Python), в которой ключами будут являться все уникальные последовательности из M букв в тексте, а значениями - списки позиций, в которых начинаются эти последовательности. После построения хеш-таблицы, осуществить в ней поиск заданного слова.

Программа должна вывести на экран через запятую по порядку позиции всех вхождений слова в текст. Если заданное слово не встречается в тексте, то вывести -1.

**Sample Input:**

cat

flkcatld lka ca otpapcatknkl camkjdlskdcat

**Sample Output:**

3,21,39

```python
def hash_table_construct(N, M):
    lst_hash = {}
    len_N = len(N)
    len_M = len(M)
    j = len_N - len_M + 1
    for i in range(j):
        slice_lst = N[i:i + len_M]
        if slice_lst in lst_hash:
            lst_hash[slice_lst].append(i)
        else:
            lst_hash[slice_lst] = [i]
    if M in lst_hash.keys():
        return lst_hash[M]
    else:
        return -1

M = input()
N = input()
res = hash_table_construct(N, M)
if res == -1:
    print(-1)
else:
    print(*res, sep = ",")
```
