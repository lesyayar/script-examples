### Задача 1. Использование условного оператора if

Даны три целых числа X,Y,Z, записанных в строку через пробел. Необходимо определить, существует ли треугольник с такими длинами сторон и, если существует, будет ли он прямоугольным.

Возможные варианты вывода программы: "no" (треугольник не существует), "yes-yes" (треугольник существует и он прямоугольный), "yes-no" (треугольник существует, но он не является прямоугольным).

**Sample Input:**

3 4 5

**Sample Output:**

yes-yes

```python
x, y, z = map(int, input().split())
if x + y > z and z + y > x and x + z > y:
    if x**2 + y**2 == z**2 or z**2 + y**2 == x**2 or x**2 + z**2 == y**2:
        print("yes-yes")
    else:
        print("yes-no")
else:
    print("no")
```

### Задача 2. Использование условного оператора if

Михаил работает в ботаническом саду - ухаживает за растениями. В назначенный день в сад привезли M литров жидкого удобрения, но новый поставщик поменял условия доставки: теперь нужно сразу же переливать удобрение в свои ёмкости, а тару возвращать курьеру. В спешке Михаил нашёл две ёмкости: кубическую с ребром A метров и цилиндрическую c высотой H метров и радиусом основания R метров. Помогите Михаилу быстро определить, в какую ёмкость поместится удобрение.

На вход программе подаются 4 вещественных числа: M,A,H,R через пробел.

Необходимо вывести на экран: cube, если удобрение целиком поместится в кубическую ёмкость (а в цилиндрическую не поместится); cylinder, если удобрение целиком поместится в цилиндрическую ёмкость (а в кубическую не поместится); cube-or-cylinder, если удобрение поместится как в кубическую, так и в цилиндрическую ёмкости; cube-and-cylinder, если удобрение не помещается целиком ни в одну из ёмкостей, но может быть разлито сразу по двум ёмкостям; fail, если этих ёмкостей не достаточно, чтобы целиком разлить удобрение.

**Sample Input:**

1000 0.5 1 0.7

**Sample Output:**

cylinder

```python
m , a, h, r = map(float, input().split())
m = m / 1000
cube_v = a ** 3
cylinder_v = 3.14 * (r**2) * h
if cube_v >= m and cylinder_v < m:
    print("cube")
elif cube_v < m and cylinder_v >= m:
    print("cylinder")
elif cube_v >= m and cylinder_v >= m:
    print("cube-or-cylinder")
elif cube_v + cylinder_v >= m:
    print("cube-and-cylinder")
elif cube_v + cylinder_v < m:
    print("fail")
```

### Задача 3. Использование цикла while

В клинике «Семейная клиника» проводится акция на УЗИ беременных женщин. Стажёру Анне поручено позвонить всем беременным женщинам, наблюдающимся в клинике, и сообщить об акции. Так как число пациентов клиники N может быть очень большим, необходимо помочь Анне автоматизировать процесс выбора беременных женщин среди всех пациентов.

На вход программе подаются следующие входные данные: в первой строке - число N всех пациентов клиники, далее идёт ровно N строк, содержащих следующую информацию через пробел - порядковый номер пациента (1, 2, 3,...), его пол (M - мужчины, F - женщины) и наличие беременности (1 - беременна, 0 - не беременна).

Программа должна вывести через запятую порядковые номера всех беременных женщин. Если беременных нет, то вывести 0.

**Sample Input:**

5

1 F 1

2 M 0

3 F 0

4 M 0

5 F 1

**Sample Output:**

1,5

```python
N = int(input())
i = 0
lst = []
while i != N:
    p = list(input().split(" "))
    if p[-1] == "1":
        i += 1
        lst.append(p[0]) 
    else: 
        i+=1        
while lst == []:
    lst.append("0")
    break
print(*lst, sep = ",")
```

### Задача 4. Использование цикла for

Хорошо известно, что в производственной лаборатории должны контролироваться температура и относительная влажность воздуха. Для измерения климатических параметров лаборатории часто используются термогигрометры, а результаты измерений заносятся в специальный журнал. Обычно температура в лаборатории колеблется от 15 до 27 градусов, а влажность воздуха изменяется от 15 до 75 процентов. Необходимо на основании всех сделанных за день записей определить среднесуточную температуру и среднесуточную влажность воздуха в лаборатории. В течение каждого дня может быть сделано разное количество записей, но всегда есть хотя бы одна запись температуры и хотя бы одна запись влажности.

На вход программе через пробел подаются зарегистрированные записи климатических измерений. Пример записи температуры: t=21C, влажности: h=52%. Все значения температур и влажностей заранее округлены до целых значений.

Необходимо вычислить среднесуточную температуру N и среднесуточную влажность воздуха M и вывести результат в формате T=N;H=M (без пробелов).

**Sample Input:**

t=21C t=22C h=52% t=18C t=19C h=60% h=60% t=17C h=56%

**Sample Output:**

T=19.4;H=57.0

```python
lst = list(input().split(" "))
N = []
M = []
for i in lst:
    if i[0] == "t":
           N.append(int(i[2:4]))
    elif i[0] == "h":
           M.append(int(i[2:4]))
n = sum(N) / len(N)
m = sum(M) / len(M)
print(f"T={n};H={m}")
```
